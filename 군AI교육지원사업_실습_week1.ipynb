## 02-02. Model Implement

+ 선형 회귀(Linear Regression) 모델의 함수는 아래와 같이 표현할 수 있습니다.

**$$ f_{w,b}(x^{(i)}) = wx^{(i)} + b \tag{1}$$**

+ 해당 공식은 **직선**을 표현하는 공식으로, $w$와 $b$의 값에 따라 직선의 모양이 달라집니다.

<br><br><br>
**Thinking : $w$와 $b$에 어떤 값을 넣으면 좋을지 직관적으로 값을 정해봅시다.**

```python
w = -2.0
b = 6.0

print(f"w: {w}")
print(f"b: {b}")
```

이제,$f_{w,b}(x^{(i)}) = wx^{(i)} + b$의 구체적인 값을 구할 수 있게 되었습니다.

>for, $x^{(0)}$, `f_wb = w * X[0] + b`
>
>for, $x^{(1)}$, `f_wb = w * X[1] + b`


```python
f_wb_0 = w * X[0] + b
f_wb_1 = w * X[1] + b

print(f'f_wb_0: {f_wb_0}')
print(f'f_wb_1: {f_wb_1}')
```

데이터의 개수가 많아지면, 하나하나 구할 수 없습니다.

이때, **함수**를 정의해서 쉽게 값을 구할 수 있도록, `compute_model_output()`함수를 정의합니다.

```python
def compute_model_output(x, w, b):
    """
    x: Data, m개의 samples
    w: model의 기울기
    b: model의 y절편
    """
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
        
    return f_wb
```

```python
f_wb = compute_model_output(X, w, b)

print(f'모델 output: {f_wb}')
```

### 03-01. Cost
+ `cost`란, **모델이 얼마나 잘 예측했는지 측정하는 척도**입니다.
<br><br><br>

비용함수의 식은 다음과 같습니다.
  $$J(w,b) = \frac{1}{2m} \sum\limits_{i = 0}^{m-1} (f_{w,b}(x^{(i)}) - y^{(i)})^2 \tag{1}$$ 
 
여기서,
  $$f_{w,b}(x^{(i)}) = wx^{(i)} + b \tag{2}$$
입니다.


- $f_{w,b}(x^{(i)})$은 앞서 구했던 $i$번째 예측값입니다.
- $(f_{w,b}(x^{(i)}) -y^{(i)})^2$은 예측값과 실제값의 차의 제곱입니다.
- 이 차이를 모두 더하고($\sum$) `2m`만큼 나누어, 평균을 구합니다.
    + 예측값과 실제값의 차이에 대한 제곱의 평균
    + 평균 오차 제곱 => `Mean Square Error(MSE)`




![image](https://user-images.githubusercontent.com/28241676/232964348-71aefaff-990e-4ee5-8f5e-59181241751c.png)

```python
m = len(X)
cost_sum = 0

for i in range(m):
    x_i, y_i = X[i], Y[i]
    f_wb_i = w * x_i + b

    cost_i = (f_wb_i - y_i)**2
    cost_sum += cost_i
    print(f'{i}번째 샘플의 Cost : {cost_i}')

print(f'전체 데이터의 Cost 평균 : {(cost_sum / (2*m))}')
```

### 03-02. Cost Function

데이터의 개수가 많아지면, 하나하나 구할 수 없습니다.

이때, **함수**를 정의해서 쉽게 값을 구할 수 있도록, `compute_cost()`함수를 정의합니다.

```python
def compute_cost(x, y, w, b): 
    """   
    Args:
      x (ndarray (m,)): 입력값 
      y (ndarray (m,)): 실제 출력값
      w,b (scalar)    : 모델의 기울기(w), 절편(b)  
    """
    # number of training examples
    m = x.shape[0] 
    
    cost_sum = 0 
    for i in range(m): 
        f_wb = w * x[i] + b   
        cost = (f_wb - y[i]) ** 2  
        cost_sum += cost 
    total_cost = (1 / (2 * m)) * cost_sum  

    return total_cost
```

```python
total_cost = compute_cost(X, Y, w, b)

print(f'총 Cost: {total_cost}')
```

**모델의 성능을 높이기 위해서, 앞서 구했던 Cost를 최소화 해야합니다.**

## 04-01. Gradient Descent

예측을 위한 선형 모델 $f_{w,b}(x^{(i)})$:

$$f_{w,b}(x^{(i)}) = wx^{(i)} + b \tag{1}$$
<br>
<br>
예측값에 대한 오차(Cost) $J(w,b)$ :
$$J(w,b) = \frac{1}{2m} \sum\limits_{i = 0}^{m-1} (f_{w,b}(x^{(i)}) - y^{(i)})^2\tag{2}$$ 
<br>
<br>
해당 오차를 최소화 하기위해, 경사하강법(Gradient Descent)를 적용합니다. :
$$ w_{new} = w_{old} - \alpha \frac{\partial J(w,b)}{\partial w} $$
$$ b_{new} = b_{old} -  \alpha \frac{\partial J(w,b)}{\partial b} $$
<br>
<br>
여기서 수치미분에 의해, $ \frac{\partial J(w,b)}{\partial w} $와 $\frac{\partial J(w,b)}{\partial b}$는 다음과 같이 계산됩니다.
$$
\begin{align}
\frac{\partial J(w,b)}{\partial w}  &= \frac{1}{m} \sum\limits_{i = 0}^{m-1} (f_{w,b}(x^{(i)}) - y^{(i)})x^{(i)} \tag{4}\\
  \frac{\partial J(w,b)}{\partial b}  &= \frac{1}{m} \sum\limits_{i = 0}^{m-1} (f_{w,b}(x^{(i)}) - y^{(i)}) \tag{5}\\
\end{align}
$$

데이터의 개수가 많아지면, 하나하나 구할 수 없습니다.

이때, 함수를 정의해서 쉽게 값을 구할 수 있도록, `compute_gradient()`함수를 정의합니다.

```python
def compute_gradient(x, y, w, b): 
    """   
    Args:
      x (ndarray (m,)): 입력값 
      y (ndarray (m,)): 실제 출력값
      w,b (scalar)    : 모델의 기울기(w), 절편(b)  
    """
    # Number of training examples
    m = x.shape[0]    
    dj_dw = 0
    dj_db = 0
    
    for i in range(m):  
        f_wb = w * x[i] + b 
        dj_dw_i = (f_wb - y[i]) * x[i] 
        dj_db_i = f_wb - y[i] 
        dj_db += dj_db_i
        dj_dw += dj_dw_i 
    dj_dw = dj_dw / m 
    dj_db = dj_db / m 
        
    return dj_dw, dj_db
```

```python
dj_dw, dj_db = compute_gradient(X, Y, w , b)

print(f'w의 기울기 : {dj_dw}')
print(f'b의 기울기 : {dj_db}')
```

## 04-02. Optimization

계산한 기울기를 바탕으로 모델의 기울기(w)와 절편(b)을 갱신해 봅시다.

```python
alpha = 0.05

dj_dw, dj_db = compute_gradient(X, Y, w , b)

w_new = w - (alpha * dj_dw)
b_new = b - (alpha * dj_db)

print(f'new w : {w_new}')
print(f'new b : {b_new}')
```

```python
f_wb_temp = compute_model_output(X, w_new, b_new)

plt.scatter(X, Y, marker='x', c='r', label='Actual Values')
plt.plot(X, f_wb_temp, c='b', label='Our Prediction')
plt.title('Y = f(X)')
plt.ylabel('Y')
plt.xlabel('X')
plt.legend()
plt.show()
```

반복문을 통해서, 최적의 w와 b 값으로 수렴시켜 봅시다.

```python
def optimization(x, y, w_init, b_init, alpha, epochs, cost_function, gradient_function): 
    J_history = []
    G_history = []
    w = w_init
    b = b_init
    
    for e in range(epochs):
        # Cost 계산
        cost = cost_function(x, y, w, b)
        # Gradient 계산
        dj_dw, dj_db = gradient_function(x, y, w , b)     

        # 파라미터 갱신(with. 경사하강법)
        b = b - alpha * dj_db                            
        w = w - alpha * dj_dw                            

        # 특정 epoch마다 학습 추이 확인
        if (e % 100) == 0:
            J_history.append(cost)
            G_history.append([w,b])
            print(f"Iteration {e:4}: Cost {J_history[-1]:0.4f} ",
                  f"dj_dw: {dj_dw: 0.5f}, dj_db: {dj_db: 0.5f}  ",
                  f"w: {w: 0.4f}, b:{b: 0.4f}")
 
    return w, b, J_history, G_history
```
